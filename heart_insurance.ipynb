#### 12. Assuming the column ‘charges’ is your target for your regression analysis, feature select the best ‘numerical_continuous’ columns using the backward elimination method.

```python
nc = df_all._get_numeric_data()
nc.dropna(how="any", inplace=True)

import statsmodels.api as sm

X = nc.drop(columns="charges")
y = nc.charges
X_1 = sm.add_constant(X)
model = sm.OLS(y,X_1).fit()
model.pvalues
```

```python
cols = list(X.columns)
pmax = 1
while len(cols)>0:
    p = []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:], index=cols)
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if pmax>0.06:
        cols.remove(feature_with_p_max)
    else:
        break
    
selected_features = cols
print(selected_features)
```

### 13. Isolate all the categorical column names into a list named ‘categorical’.

```python
categorical = df_all._get_numeric_data().drop(columns="heart disease")
categorical.info()
```

### 14. Assuming the column ‘heart_disease’ is the target for your classification analysis, run a chi contingency test to identify the best categorical and numerical_discrete features to proceed with the analysis.

```python
from scipy.stats import chi2_contingency

chi_res = chi2_contingency(pd.crosstab(df_all['heart disease'], df_all['age']))
print('Chi2 Statistic: {}, p-value: {}'.format(chi_res[0], chi_res[1]))
```

```python
chi2_check = []
for i in categorical:
    if chi2_contingency(pd.crosstab(df_all['heart disease'], df_all[i]))[1] < 0.05:
        chi2_check.append('Reject Null Hypothesis')
    else:
        chi2_check.append('Fail to Reject Null Hypothesis')
res = pd.DataFrame(data = [categorical, chi2_check] 
             ).T 
res.columns = ['Column', 'Hypothesis']
print(res)
```

### 15. Using ColumnTransformer, OneHotEncode the categorical columns.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

oe = OneHotEncoder()
ct = ColumnTransformer([('encoder', oe, [0])], remainder='passthrough')

categorical = np.array(ct.fit_transform(categorical), dtype=np.str)
categorical
```
